```python


# Generate between 200 and 1000 order items (1-5 items per order)
num_orders = 200
records = []
order_item_counts = [random.randint(1, 5) for _ in range(num_orders)]
total_items = sum(order_item_counts)

# Used to track products per order to avoid duplicates
order_products = {}

for order_id in range(1, num_orders + 1):
    # Determine how many items in this order
    num_items = order_item_counts[order_id - 1]
    
    # Select distinct products for this order
    product_ids = random.sample(range(1, 201), num_items)
    order_products[order_id] = product_ids
    
    for product_id in product_ids:
        # Generate quantity (mostly 1-3, occasionally more)
        quantity = random.randint(1, 3) if random.random() < 0.8 else random.randint(4, 10)
        
        # Generate realistic price (based on product category ranges)
        category_based_price = {
            "Electronics": random.uniform(5000, 150000),
            "Footwear": random.uniform(2000, 30000),
            "Books": random.uniform(500, 10000),
            "Fashion": random.uniform(1500, 25000),
            "Home & Garden": random.uniform(1000, 50000),
            "Beauty": random.uniform(800, 20000)
        }
        price = round(random.choice(list(category_based_price.values())), 2)
        
        # Generate discount (70% no discount, 20% small discount, 10% large discount)
        if random.random() < 0.7:
            discount = 0.00
        elif random.random() < 0.9:
            discount = round(price * random.uniform(0.05, 0.2), 2)
        else:
            discount = round(price * random.uniform(0.25, 0.5), 2)
        
        # Generate item attributes
        attributes = {}
        
        # Common attributes
        if random.random() < 0.4:
            attributes["warranty"] = random.choice(["6 months", "1 year", "2 years"])
        if random.random() < 0.3:
            attributes["bundle_offer"] = True
        
        # Size-based attributes
        if random.random() < 0.3:
            attributes["size"] = random.choice(["S", "M", "L", "XL"])
        
        # Color attributes
        if random.random() < 0.25:
            attributes["color"] = random.choice(["Black", "White", "Red", "Blue", "Green"])
        
        # Special attributes
        if random.random() < 0.15:
            special_attrs = {
                "RAM_upgrade": random.choice(["16GB", "32GB", "64GB"]),
                "storage_upgrade": random.choice(["512GB SSD", "1TB SSD", "2TB HDD"]),
                "gift_wrap": True,
                "engraving": random.choice(["Initials", "Message", "Date"]),
                "installation": random.choice(["Basic", "Premium", "Professional"])
            }
            # Fix: add a single random special attribute key-value pair
            key, value = random.choice(list(special_attrs.items()))
            attributes[key] = value
        
        # Escape single quotes in JSON
        attr_str = json.dumps(attributes).replace("'", "''")
        
        records.append(
            f"({order_id}, {product_id}, {quantity}, {price}, {discount}, '{attr_str}')"
        )

# Generate SQL
sql = "INSERT INTO order_items (order_id, product_id, quantity, price, discount, item_attributes)\nVALUES\n"
sql += ",\n".join(records) + ";"

print(sql)
```

```python
-- Find customers who have placed orders over $1000.
-- (Query: Get names and emails of customers with orders over $1000)
SELECT name, email
FROM customers c
WHERE customer_id IN (
    SELECT customer_id
    FROM orders
    WHERE total_amount > 1000
);

-- Find all products above the average product price
SELECT product_id, product_name, price
FROM products p
WHERE price > (
    SELECT AVG(price) FROM products
);

-- (Optional: Get the average product price)
-- SELECT AVG(price) FROM products;



-- Get the total number of orders per customer, but only for customers who have made at least 2 orders
SELECT customer_id, total_orders
FROM (
    SELECT customer_id, COUNT(*) AS total_orders
    FROM orders o
    GROUP BY customer_id
) AS customer_order_counts
WHERE total_orders >= 2;


-- Example of a Common Table Expression (CTE)
-- (Template)
-- WITH cte_name AS (
--     SELECT ...
-- )
-- SELECT ... FROM cte_name;

-- Find products priced above the average product price using a CTE
WITH avg_price AS (
    SELECT AVG(price) AS avg_price
    FROM products
)
SELECT product_id, product_name, price
FROM products p, avg_price a
WHERE p.price > a.avg_price;

-- Find top 3 customers by total spending using CTEs
WITH customer_spending AS (
    SELECT customer_id, SUM(total_amount) AS total_spent
    FROM orders
    GROUP BY customer_id
),
top_customers AS (
    SELECT customer_id, total_spent
    FROM customer_spending
    ORDER BY total_spent DESC
    LIMIT 3
)
SELECT c.name, tc.total_spent
FROM customers c
JOIN top_customers tc ON c.customer_id = tc.customer_id
ORDER BY total_spent DESC;
```
